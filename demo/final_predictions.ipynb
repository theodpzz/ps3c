```python
import numpy as np
import pandas as pd
```

```python
path_eval = "./probabilities_eval.csv"

predictions_eval_csv = pd.read_csv(path_eval)

# prediction on rubbish vs non-rubbish
predictions_eval_csv['rubbish_prediction'] = 1*(predictions_eval_csv.rubbish >= 0.46)

# prediction healthy vs unhealthy
predictions_eval_csv['healthy_prediction'] = 1*(predictions_eval_csv.healthy > predictions_eval_csv.unhealthy)

# final prediction: first assign rubbish, then healthy or unhealthy
predictions_eval_csv['final_prediction'] = np.where(
    predictions_eval_csv['rubbish_prediction'] == 1, 
    'rubbish',
    np.where(predictions_eval_csv['healthy_prediction'] == 1, 'healthy', 'unhealthy')
)

predictions_eval_csv[["image_name", "final_prediction"]]
```
